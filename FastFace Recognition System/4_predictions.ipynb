```python
import pandas 
import numpy 
```

```python
## Extract Data from database 
name='academy:register'
retrieve_dict=face_reco.r.hgetall(name)
retrieve_series=pandas.Series(retrieve_dict)
retrieve_series=retrieve_series.apply(lambda x:numpy.frombuffer(x,dtype=numpy.float32))
index=retrieve_series.index
index=list(map(lambda x:x.decode(),index))
retrieve_series.index=index
retrieve_df=retrieve_series.to_frame().reset_index()
retrieve_df.columns=['name_role','Facial Features']
retrieve_df[['Name','Role']]=retrieve_df['name_role'].apply(lambda x:x.split('@')).apply(pandas.Series)
retrieve_df
```

### Real Time Predictions

```python
cap=cv2.VideoCapture(0)


while True:
    ret,frame=cap.read()
    if ret==False:
        break
    
    
    pred_frame=face_reco.face_prediction(frame,retrieve_df,'Facial Features',['Name','Role'],thresh=0.5) 
    
    cv2.imshow('prediction',pred_frame)
    
    
    if cv2.waitKey(1) == 27:
        break 
        

        
cv2.destroyAllWindows()
cap.release()
```
